```python
import numpy as np
import cv2
import math
print('problem 1')
img1= cv2.imread('O:\\Photographs\\Resnet\\Car_1.jpg')
img1=cv2.resize(img1, (0,0), fx=0.3, fy=0.3)
a1=0.02; b1=0.01; c1=10; d1=0.01; e1=-0.02; f1=5

height, width, depth = img1.shape
#height, width = img1.shape
def warpImg(imgCv, a, b, c, d, e, f):
    
    if(imgCv.ndim == 3): 
        h, w, depth = imgCv.shape
    elif(imgCv.ndim == 2):
        h, w = imgCv.shape
    
    M_fwd = np.asarray([[1+a, b, c],
                        [d, 1+e, f],
                        [0,    0,   1]])
    Vert0 = M_fwd.dot(np.asarray([[0,0,1]]).T)
    Vert1 = M_fwd.dot(np.asarray([[w-1,0,1]]).T)
    Vert2 = M_fwd.dot(np.asarray([[w-1, h-1,1]]).T)
    Vert3 = M_fwd.dot(np.asarray([[0, h-1,1]]).T)
    #print([Vert0, Vert1, Vert2, Vert3])
    
    #x_lower  = min([Vert0[0],Vert1[0],Vert2[0],Vert3[0]])
    x_higher = max([Vert0[0],Vert1[0],Vert2[0],Vert3[0]])
    #y_lower  = min([Vert0[1],Vert1[1],Vert2[1],Vert3[1]])
    y_higher = max([Vert0[1],Vert1[1],Vert2[1],Vert3[1]])
    
    if(imgCv.ndim == 3): 
        newImg = np.empty([int(math.ceil(y_higher)), 
                       int(math.ceil(x_higher)),3 ])
        M_rev = np.linalg.inv(M_fwd)
        print(M_fwd)
        #print([x_higher, y_higher])
        counter = 0

        for y in range (0, int(math.floor(y_higher))):
            counter = 0
            for x in range(0, int(math.floor(x_higher))):
                counter+=1
                Ori = M_rev.dot(np.asarray([[x, y, 1]]).T)
                x_o = Ori[0]; y_o = Ori[1];
                if y_o > h-1 or x_o > w-1:
                    break
                #print([x_o,y_o])   
                w_left = x_o - math.floor(x_o); w_up = y_o - math.floor(y_o)
                intensity_up = w_left*imgCv[math.floor(y_o),
                    math.ceil(x_o), :] + (1-w_left)*imgCv[math.floor(y_o), math.floor(x_o), :]
                intensity_down = w_left*imgCv[math.ceil(y_o),
                    math.ceil(x_o), :] + (1-w_left)*imgCv[math.ceil(y_o), math.floor(x_o), :]

                intensity = w_up*intensity_down + (1-w_up)*intensity_up

                newImg[y,x,:] = np.uint8(np.floor(intensity))
                
    elif(imgCv.ndim == 2):
        newImg = np.empty([int(math.ceil(y_higher)), 
                       int(math.ceil(x_higher))])
        M_rev = np.linalg.inv(M_fwd)
        #print(M_fwd)
        #print([x_higher, y_higher])
        counter = 0

        for y in range (0, int(math.floor(y_higher))):
            counter = 0
            for x in range(0, int(math.floor(x_higher))):
                counter+=1
                Ori = M_rev.dot(np.asarray([[x, y, 1]]).T)
                x_o = Ori[0]; y_o = Ori[1];
                if y_o > h-1 or x_o > w-1:
                    break
                #print([x_o,y_o])   
                w_left = x_o - math.floor(x_o); w_up = y_o - math.floor(y_o)
                intensity_up = w_left*imgCv[math.floor(y_o),
                    math.ceil(x_o)] + (1-w_left)*imgCv[math.floor(y_o), math.floor(x_o)]
                intensity_down = w_left*imgCv[math.ceil(y_o),
                    math.ceil(x_o)] + (1-w_left)*imgCv[math.ceil(y_o), math.floor(x_o)]

                intensity = w_up*intensity_down + (1-w_up)*intensity_up

                newImg[y,x] = np.uint8(np.floor(intensity))

    
  
    """newImg = np.empty([int(math.ceil(height+max([a,0])*width+max([b,0])*height+c)), 
                       int(math.ceil(width+max([d,0])*width+max([e,0])*height+f))]) """
    
    """ newImg = np.empty([int(math.ceil(height+max([d,0])*width+max([e,0])*height+f)), 
                       int(math.ceil(width+max([a,0])*width+max([b,0])*height+c)),
                       3 ])
    for i in range(height):
        for j in range(width):
            
            i_2 = int(math.ceil(i+d*j+e*i+f))
            j_2 = int(math.ceil(j+a*j+b*i+c))
            newImg[i_2, j_2, :] = imgCv[i, j, :]
            #newImg[i_2, j_2] = imgCv[i, j]"""
   
    return newImg


img2 = warpImg(img1, a1, b1, c1, d1, e1, f1)
h1, w1, d1=img1.shape; h2, w2, d2 = img2.shape
#h1, w1=img1.shape

img2 = cv2.convertScaleAbs(img2)


img1_img2_overlap = 0.5*img1[:h2, :, :] + 0.5*img2[:h1,: w1, :]
img1_img2_diff = (np.int16(img1[:h2, :, :])  + 255 - np.int16(img2[:, :w1, :])) / 2
#img1_img2_overlap = 0.5*img1 + 0.5*img2[:h1, :w1]
img1_img2_concat = np.concatenate((img1, img2[:h1, :w1, :]), axis=0)

title = 'Combinition of difference'
cv2.imshow(title, cv2.convertScaleAbs(img1_img2_diff))
cv2.waitKey(0)
cv2.destroyAllWindows()

print('problem 2')

"""kernel_x = np.array([[-1., 1.], [-1., 1.]])
kernel_y = np.array([[-1., -1.], [1., 1.]])
kernel_t = np.array([[1., 1.], [1., 1.]])
"""

#crop the image
h2, w2, d2 = img2.shape


img1_bw = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_bw = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

#Take in a Gray Scale Image and blur it using gaussian blur filter
def GaussianBlur_55(img):
    
    Gaussian_5 = np.asarray([1, 4, 6, 4, 1])/16
    
    h_g, w_g = img.shape
    
    new_img = np.empty([h_g, w_g])
    
    #zero padding
    img = np.concatenate(( np.zeros([h_g, 2]), img), axis = 1)
    img = np.concatenate((img,  np.zeros([h_g, 2])), axis = 1)
    img = np.concatenate((np.zeros([2, w_g+4]), img), axis = 0)
    img = np.concatenate((img, np.zeros([2, w_g+4])), axis = 0) 
    
    for y in range (h_g):
            
        for x in range (w_g):
            #print(x)
            blurred_intensity = np.inner(Gaussian_5, img[y+2, x:x+5])
            new_img[y, x] = blurred_intensity
    
    for y in range (h_g):
             
        for x in range (w_g):
            blurred_intensity = np.inner(Gaussian_5, img[y:y+5, x+2])
            new_img[y, x] = blurred_intensity
            
    return new_img[2:-2, 2:-2]

#input two Gray Scale images and output the 6 linear motion params
def motion_est(img_ori, img_new, scale):
    
    height, width = img_ori.shape
    x_lower = int(math.floor(width*scale*0.01)) ; x_higher = int(math.floor(width-x_lower)); y_lower = int(math.floor(height*scale*0.01)); y_higher = int(math.floor(height-y_lower))

    A_T_A = np.zeros([6,6])
    A_T_B = np.zeros([1,6])
    for y in range(y_lower+1, y_higher-1):
        for x in range (x_lower+1, x_higher-1):

            i_x = np.int64(img_ori[y, x+1]) - np.int64(img_ori[y,x-1])
            i_y = np.int64(img_ori[y+1, x]) - np.int64(img_ori[y-1,x])

            i_t = np.int64(img_new[y, x]) - np.int64(img_ori[y,x])

            tmp_A = []; 
            tmp_A.append([(i_x**2)*x**2, (i_x**2)*x*y, (i_x**2)*x, i_x*i_y*x**2, i_x*i_y*x*y, i_x*i_y*x])
            tmp_A.append([(i_x**2)*x*y, (i_x**2)*y**2, (i_x**2)*y, i_x*i_y*x*y, i_x*i_y*y**2, i_x*i_y*y])
            tmp_A.append([(i_x**2)*x, (i_x**2)*y, (i_x**2), i_x*i_y*x, i_x*i_y*y, i_x*i_y])
            tmp_A.append([i_x*i_y*x**2, i_x*i_y*x*y, (i_x*i_y)*x, (i_y**2)*x**2, (i_y**2)*x*y, (i_y**2)*x])
            tmp_A.append([i_x*i_y*x*y, i_x*i_y*y**2, i_x*i_y*y, (i_y**2)*x*y, (i_y**2)*y**2, (i_y**2)*y])
            tmp_A.append([i_x*i_y*x, i_x*i_y*y, i_x*i_y, (i_y**2)*x, (i_y**2)*y, i_y**2])

            tmp_A = np.asarray(tmp_A)

            A_T_A += tmp_A
            A_T_B += -np.asarray([[i_t*i_x*x, i_t*i_x*y, i_t*i_x, i_t*i_y*x, i_t*i_y*y, i_t*i_y]])

    q = np.linalg.inv(A_T_A).dot(A_T_B.T)[:,0]
    print("estimators:" + str(q))
    print("scale:" + str(scale))
    return (q[0], q[1], q[2], q[3], q[4], q[5])
    


#The pyramid has its most downsample image at the end
img1_pyramid = []
img2_pyramid = []
img1_bw_blurred = img1_bw; img2_bw_blurred = img2_bw

while(img1_bw_blurred.shape[0]>32):
    img1_bw_blurred = GaussianBlur_55(img1_bw_blurred); img2_bw_blurred = GaussianBlur_55(img2_bw_blurred)
    img1_bw_blurred = img1_bw_blurred[::2, ::2]; img2_bw_blurred = img2_bw_blurred[::2, ::2]
    
    img1_pyramid.append(img1_bw_blurred)
    img2_pyramid.append(img2_bw_blurred)


a, b, c, d, e, f = motion_est(img1_pyramid[-1], img2_pyramid[-1], 1)

c *= 2; f *= 2
index = -2

while True:
    
  
    
    img1_tmp = img1_pyramid[index]; img2_tmp = img2_pyramid[index]
    
    img1_warp_tmp = warpImg(img1_tmp, a, b, c, d, e, f)
    
    a_t, b_t, c_t, d_t, e_t, f_t = motion_est(img1_warp_tmp, img2_tmp, -index)
    
    a += a_t; b += b_t; c += c_t; d += d_t; e += e_t; f += f_t;
    
    c *= 2; f*=2
    
    print("Aggregated estimators: " + str([a,b,c,d,e,f]))
    
    index += -1
    if index < -len(img1_pyramid):
        break
    
    
cv2.imshow("blurred", cv2.convertScaleAbs(img1_bw_blurred))
cv2.resize(img1, (0,0), fx=0.3, fy=0.3)
cv2.waitKey(0)
cv2.destroyAllWindows()



cv2.imshow(title, cv2.convertScaleAbs(img2[y_lower:y_higher, x_lower:x_higher, :]))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import tensorflow as tf
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
import glob
print('3333')
model = ResNet50(weights='imagenet')

path = "O:\\Photographs\\RedNet"
images_path = glob.glob(path + "\\*.jpg")


for img_path in images_path:
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    
    preds = model.predict(x)
    # decode the results into a list of tuples (class, description, probability)
    # (one such list for each sample in the batch)
    print('Predicted:', decode_predictions(preds, top=3)[0])
```
