```python
#Sturm_Triggs_1
def normalize_Keypts(pts2d):
    mean = np.mean(pts2d, axis=0)
    pts2d = pts2d - mean
    var_mean = np.mean(list(map(np.linalg.norm, pts2d)))
    return pts2d/var_mean*np.sqrt(2)

#Sturm_Triggs_3
def est_ScaleFactor(pt2d_1, pt2d_2, epipole, F, scale_prev):
    #pt2d_1 = np.concatenate((pt2d_1, np.ones(len(pt2d_1))), axis=1)
    #pt2d_2 = np.concatenate((pt2d_2, np.ones(len(pt2d_2))), axis=1)
    pt2d_1 = np.concatenate((pt2d_1, [1]))
    pt2d_2 = np.concatenate((pt2d_2, [1]))
    return (np.inner(np.cross(epipole, pt2d_2), 
             F.dot(pt2d_1))*scale_prev
     /np.linalg.norm(np.cross(epipole, pt2d_1)))


#print(normalize_Keypts(corres_Query))
```

```python
#Calculate a fundamental matrix from the 8 corresponding points
def camMat_aus8Pts(pts2d_1, pts2d_2):
    A = []
    for i in range(min(len(pts2d_1), len(pts2d_2))):
        x_l = pts2d_1[i, 0]; x_r = pts2d_2[i,0];
        y_l = pts2d_1[i, 1]; y_r = pts2d_2[i,1];
        A.append([x_l*x_r, x_l*y_r, x_l, y_l*x_r, y_l*y_r, y_l, x_r, y_r, 1])
    
    A = np.asarray(A)
    u_A, s_A, v_A = np.linalg.svd(A)
    q = v_A.T[:, -1]
    F=np.asarray([q[0:3], q[3:6], q[6:]]).T
    return F

#Estimate the error for RANSAC metric
def error_F_RANSAC(pt2d_1, pt2d_2, F):
    return (np.concatenate((pt2d_1, [1])).T
         .dot(F)
         .dot(np.concatenate((pt2d_2, [1]))))

#Determine if the RANSAC random picking has no outlier involved
def hasOutlier_RANSAC(pts2d_1, pts2d_2, F):
    hasOutlier = False
    num_Outlier = 0
    for i in range(len(pts2d_1)):
        if abs(error_F_RANSAC(pts2d_1[i], pts2d_2[i], F)) > 0.1:
            num_Outlier+=1
            hasOutlier = True
    #print(num_Outlier)
    return hasOutlier

#Determine a RANSAC estimated Fundamental Matrix from large number of 2d point correspondences
def est_F(matchPts1, matchPts2):
    counter = 0
    hasFoundF = False
    matchPts1 = normalize_Keypts(matchPts1)
    matchPts2 = normalize_Keypts(matchPts2)
    while(counter <= 10000 and (not hasFoundF)):
        index_8 = np.random.choice(a=range(len(matchPts1)), size = 8, replace = False)
        #print(index_8)
        F_this = camMat_aus8Pts(matchPts1[index_8], matchPts2[index_8])
        hasFoundF = not hasOutlier_RANSAC(matchPts1[index_8], matchPts2[index_8], F_this)
        counter += 1
    print('RANSAC iterations' + str(counter))
    U, S, V = np.linalg.svd(F_this)
    e = V.T[:, -1]
    return (F_this, e, matchPts1[index_8], matchPts2[index_8], counter)

#Estimate a Fundamental Matrix from the Keypoint matches
#After estimating F, use F to filter out Outliers/Errors in the keypoint matches
def est_F_fromIdx(queryIdx, trainIdx, queryKp, trainKp):
    queryPts = []
    trainPts = []
    for q in np.asarray([queryIdx, trainIdx]).T:
        queryPts.append(list(queryKp[q[0]].pt))
        trainPts.append(list(trainKp[q[1]].pt)) 
    #print(queryPts)
    #print(trainPts)
    queryPts = np.asarray(queryPts)
    trainPts = np.asarray(trainPts)
    F, e, mPts1, mPts2, numIt  = est_F(queryPts, trainPts)
    index = filter_Matches_byF(queryPts, trainPts, F)
    retval = (F, queryIdx[index], trainIdx[index])
    return retval

def filter_Matches_byF(queryPts, trainPts, F):
    index = []
    queryPts = normalize_Keypts(queryPts)
    trainPts = normalize_Keypts(trainPts)
    for i in range(len(queryPts)):
        if error_F_RANSAC(queryPts[i], trainPts[i], F) < 200:
            index.append(i)
    return index
    
    
F_foo, e, mPts1_foo, mPts2_foo, numIt_foo = est_F(corres_Query, corres_Train)
print(mPts1_foo)

```
